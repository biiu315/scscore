```python
check_decoy=[]
for idx, i in enumerate(data):
    if 'decoy' not in i.keys():
        check_decoy.append(idx)
```

```python
data_count={}
for i in data_decoy:
    len_key=len(i.keys())
    num_decoy=len_key-1
    if f'{num_decoy}' not in data_count.keys():
        data_count[f'{num_decoy}']=[]
        data_count[f'{num_decoy}'].append(i)
    else:
        data_count[f'{num_decoy}'].append(i)
        
```
